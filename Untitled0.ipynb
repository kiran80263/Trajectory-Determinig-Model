```python
# Import necessary libraries
import cv2
import numpy as np
import matplotlib.pyplot as plt
from collections import defaultdict
import os
from datetime import datetime, timedelta
import time
from google.colab.patches import cv2_imshow
from google.colab import files
import io
from IPython.display import display, HTML, clear_output
from PIL import Image

# Check if we're in Colab
try:
    from google.colab import drive
    IN_COLAB = True
    print("Running in Google Colab")
except:
    IN_COLAB = False
    print("Not running in Google Colab")

# Install required packages if in Colab
if IN_COLAB:
    print("Installing required packages...")
    !pip install ultralytics supervision -q
    from ultralytics import YOLO
    import supervision as sv
    print("Packages installed successfully!")

# Define vehicle classes and colors
VEHICLE_CLASSES = {
    2: {'name': 'car', 'color': 'blue', 'display_color': (255, 0, 0)},     # Blue for cars
    1: {'name': 'bike', 'color': 'green', 'display_color': (0, 255, 0)},    # Green for bikes
    7: {'name': 'rickshaw', 'color': 'orange', 'display_color': (0, 165, 255)}, # Orange for rickshaws
    5: {'name': 'bus', 'color': 'red', 'display_color': (0, 0, 255)},      # Red for HTVs
    7: {'name': 'truck', 'color': 'red', 'display_color': (0, 0, 255)}       # Red for HTVs
}

class LaneChangeAnalyzer:
    def __init__(self, model_path="yolov8x.pt", confidence=0.3, lane_count=5, time_group_seconds=5):
        """
        Initialize the lane change analyzer with YOLOv8 model

        Args:
            model_path: Path to YOLOv8 model weights
            confidence: Detection confidence threshold
            lane_count: Number of lanes to divide the road into
            time_group_seconds: Number of seconds to group time frames by
        """
        self.model = YOLO(model_path)
        self.confidence = confidence
        self.tracker = sv.ByteTrack()
        self.trajectories = defaultdict(list)
        self.vehicle_types = {}
        self.lane_count = lane_count
        self.time_group_seconds = time_group_seconds

        # Create output directory for results
        self.output_dir = f"lane_change_analysis_{datetime.now().strftime('%Y%m%d_%H%M%S')}"
        os.makedirs(self.output_dir, exist_ok=True)

        # Progress tracking
        self.total_frames = 0
        self.processed_frames = 0
        self.start_time = None
        self.fps = 0

    def detect_lane(self, x_position, frame_width):
        """
        Determine which lane a vehicle is in based on its x-position

        Args:
            x_position: X-coordinate of the vehicle
            frame_width: Width of the video frame

        Returns:
            Lane number (0 to lane_count-1)
        """
        lane_width = frame_width / self.lane_count
        lane = int(x_position / lane_width)
        return min(max(0, lane), self.lane_count - 1)  # Ensure lane is within bounds

    def process_video(self, video_path, sample_rate=1, max_frames=None, display_output=True):
        """
        Process the video and track vehicles' lane changes

        Args:
            video_path: Path to the traffic video file
            sample_rate: Process every Nth frame (default: 1)
            max_frames: Maximum number of frames to process (default: None - process all)
            display_output: Whether to display frames during processing (default: True)
        """
        cap = cv2.VideoCapture(video_path)
        frame_width = int(cap.get(cv2.CAP_PROP_FRAME_WIDTH))
        frame_height = int(cap.get(cv2.CAP_PROP_FRAME_HEIGHT))
        self.fps = int(cap.get(cv2.CAP_PROP_FPS))

        # Calculate total frames
        self.total_frames = int(cap.get(cv2.CAP_PROP_FRAME_COUNT))
        if max_frames and max_frames < self.total_frames:
            self.total_frames = max_frames

        print(f"Video properties: {frame_width}x{frame_height}, {self.fps} FPS, {self.total_frames} frames")

        # Set up video writer for the output
        output_path = os.path.join(self.output_dir, "lane_change_analysis.mp4")
        fourcc = cv2.VideoWriter_fourcc(*'mp4v')
        out = cv2.VideoWriter(output_path, fourcc, self.fps/sample_rate, (frame_width, frame_height))

        frame_count = 0
        self.processed_frames = 0
        self.start_time = time.time()

        # Initialize lane positions
        lane_positions = []
        for i in range(self.lane_count + 1):
            lane_x = int(frame_width * i / self.lane_count)
            lane_positions.append(lane_x)

        # Initialize progress display
        progress_html = HTML(
            """<progress style="width:100%" max="100" value="0"></progress>
            <div id="progress-label">Processing: 0%</div>"""
        )
        if display_output:
            display(progress_html)

        start_time = time.time()
        last_update_time = start_time

        while cap.isOpened():
            success, frame = cap.read()
            if not success:
                break

            frame_count += 1

            # Skip frames based on sample_rate
            if frame_count % sample_rate != 0:
                continue

            # Stop if max_frames is reached
            if max_frames and self.processed_frames >= max_frames:
                break

            self.processed_frames += 1
            current_time_seconds = self.processed_frames / self.fps  # Time in video in seconds

            # Draw lane lines
            for lane_x in lane_positions:
                cv2.line(frame, (lane_x, 0), (lane_x, frame_height), (255, 255, 255), 1)

            # Run YOLOv8 inference
            results = self.model.track(frame, persist=True, conf=self.confidence, classes=[1, 2, 5, 7])

            if results[0].boxes.id is not None:
                # Get detection and tracking results
                detections = sv.Detections.from_ultralytics(results[0])
                detections = self.tracker.update_with_detections(detections)

                # Process each detection
                for i, (bbox, confidence, class_id, tracker_id) in enumerate(zip(
                    detections.xyxy, detections.confidence, detections.class_id, detections.tracker_id
                )):
                    if class_id in VEHICLE_CLASSES:
                        # Store the vehicle type for this ID
                        if tracker_id not in self.vehicle_types:
                            self.vehicle_types[tracker_id] = class_id

                        # Calculate center point
                        x1, y1, x2, y2 = map(int, bbox)
                        center_x = (x1 + x2) // 2
                        center_y = (y1 + y2) // 2

                        # Determine current lane
                        current_lane = self.detect_lane(center_x, frame_width)

                        # Append to trajectory with lane information and timestamp
                        self.trajectories[tracker_id].append((self.processed_frames, center_x, center_y, current_lane, current_time_seconds))

                        # Draw bounding box and ID
                        color = VEHICLE_CLASSES[class_id]['display_color']
                        cv2.rectangle(frame, (x1, y1), (x2, y2), color, 2)
                        cv2.putText(frame, f"ID:{tracker_id} Lane:{current_lane+1}", (x1, y1-10),
                                    cv2.FONT_HERSHEY_SIMPLEX, 0.5, color, 2)

            # Save the frame
            out.write(frame)

            # Update progress display every second
            current_time = time.time()
            if current_time - last_update_time >= 1.0 or self.processed_frames == self.total_frames:
                progress_percent = min(100, int((self.processed_frames / self.total_frames) * 100))
                elapsed_time = current_time - start_time
                frames_per_second = self.processed_frames / elapsed_time if elapsed_time > 0 else 0
                eta = (self.total_frames - self.processed_frames) / frames_per_second if frames_per_second > 0 else 0

                progress_html.value = f"""<progress style="width:100%" max="100" value="{progress_percent}"></progress>
                <div id="progress-label">Processing: {progress_percent}% | FPS: {frames_per_second:.2f} | ETA: {eta:.2f}s</div>"""

                if display_output and self.processed_frames % 10 == 0:
                    # Display the frame in Colab
                    clear_output(wait=True)
                    display(progress_html)
                    rgb_frame = cv2.cvtColor(frame, cv2.COLOR_BGR2RGB)
                    plt.figure(figsize=(12, 8))
                    plt.imshow(rgb_frame)
                    plt.axis('off')
                    plt.show()

                last_update_time = current_time

        cap.release()
        out.release()

        # Display final statistics
        total_time = time.time() - start_time
        print(f"Processing complete: {self.processed_frames} frames in {total_time:.2f} seconds ({self.processed_frames/total_time:.2f} FPS)")
        print(f"Detected {len(self.vehicle_types)} unique vehicles")

        # After processing, analyze lane changes and generate plots
        self.analyze_lane_changes(frame_width, frame_height)

        # In Colab, download the results
        if IN_COLAB:
            self._download_results()

    def analyze_lane_changes(self, frame_width, frame_height):
        """
        Analyze lane changes and generate plots

        Args:
            frame_width: Width of the video frame
            frame_height: Height of the video frame
        """
        print("Analyzing lane changes and generating plots...")

        # Prepare data for plotting
        lane_changes = {
            'car': [],
            'bike': [],
            'rickshaw': [],
            'htv': []
        }

        # Detect lane changes in trajectories
        for track_id, points in self.trajectories.items():
            if track_id not in self.vehicle_types or len(points) < 10:
                continue

            vehicle_class = self.vehicle_types[track_id]
            vehicle_type = VEHICLE_CLASSES[vehicle_class]['name']

            # Group points by lane to detect lane changes
            current_lane = points[0][3]  # Initial lane
            lane_segments = []
            segment_start = points[0]

            for point in points[1:]:
                if point[3] != current_lane:  # Lane changed
                    # Record the lane change
                    lane_segments.append((segment_start, point))
                    segment_start = point
                    current_lane = point[3]

            # Add the final segment if it exists
            if segment_start != points[-1]:
                lane_segments.append((segment_start, points[-1]))

            # Add lane changes to the appropriate category
            if vehicle_type == 'car':
                lane_changes['car'].extend(lane_segments)
            elif vehicle_type == 'bike':
                lane_changes['bike'].extend(lane_segments)
            elif vehicle_type == 'rickshaw':
                lane_changes['rickshaw'].extend(lane_segments)
            elif vehicle_type in ['bus', 'truck']:
                lane_changes['htv'].extend(lane_segments)

        # Generate plots
        print("Generating combined plot...")
        self._generate_combined_plot(lane_changes, frame_width, frame_height)

        print("Generating vehicle-specific plots...")
        self._generate_vehicle_type_plot('car', lane_changes['car'], 'blue', frame_width, frame_height)
        self._generate_vehicle_type_plot('bike', lane_changes['bike'], 'green', frame_width, frame_height)
        self._generate_vehicle_type_plot('rickshaw', lane_changes['rickshaw'], 'orange', frame_width, frame_height)
        self._generate_vehicle_type_plot('htv', lane_changes['htv'], 'red', frame_width, frame_height)

        # Generate time-based plots
        print("Generating time-based lane change plots...")
        self._generate_time_based_plot(lane_changes, frame_width)

        print("Plot generation complete!")

    def _generate_combined_plot(self, lane_changes, frame_width, frame_height):
        """
        Generate a combined plot showing all vehicle lane changes

        Args:
            lane_changes: Dictionary of lane changes by vehicle type
            frame_width: Width of the video frame
            frame_height: Height of the video frame
        """
        plt.figure(figsize=(15, 10))
        plt.title("All Vehicles Lane Change Behavior")

        # Draw lanes
        for i in range(self.lane_count + 1):
            lane_x = int(frame_width * i / self.lane_count)
            plt.axvline(x=lane_x, color='black', linestyle='--', alpha=0.3)

        # Draw lane changes for each vehicle type
        for vehicle_type, segments in lane_changes.items():
            color = {'car': 'blue', 'bike': 'green', 'rickshaw': 'orange', 'htv': 'red'}[vehicle_type]

            for start_point, end_point in segments:
                start_x, start_y = start_point[1], start_point[2]
                end_x, end_y = end_point[1], end_point[2]

                # Only draw if there's significant horizontal movement
                if abs(start_x - end_x) > frame_width * 0.02:
                    plt.arrow(start_x, start_y,
                             end_x - start_x, end_y - start_y,
                             head_width=frame_height * 0.01, head_length=frame_width * 0.01,
                             fc=color, ec=color, alpha=0.6)

        plt.xlim(0, frame_width)
        plt.ylim(frame_height, 0)  # Invert Y-axis to match video coordinates
        plt.xlabel("X-coordinate")
        plt.ylabel("Y-coordinate")
        plt.legend(['Lane Boundaries', 'Cars (Blue)', 'Bikes (Green)', 'Rickshaws (Orange)', 'HTVs (Red)'])
        plt.grid(False)
        plt.savefig(os.path.join(self.output_dir, "all_vehicles_lane_changes.png"), dpi=300)
        plt.show()  # Show in notebook

    def _generate_vehicle_type_plot(self, vehicle_type, segments, color, frame_width, frame_height):
        """
        Generate a plot showing lane changes for a specific vehicle type

        Args:
            vehicle_type: Type of vehicle
            segments: List of lane change segments
            color: Color to use for plotting
            frame_width: Width of the video frame
            frame_height: Height of the video frame
        """
        if not segments:
            print(f"No lane changes detected for {vehicle_type}.")
            return

        plt.figure(figsize=(15, 10))
        plt.title(f"{vehicle_type.title()} Lane Change Behavior")

        # Draw lanes
        for i in range(self.lane_count + 1):
            lane_x = int(frame_width * i / self.lane_count)
            plt.axvline(x=lane_x, color='black', linestyle='--', alpha=0.3)

        # Draw lane changes
        for start_point, end_point in segments:
            start_x, start_y = start_point[1], start_point[2]
            end_x, end_y = end_point[1], end_point[2]

            # Only draw if there's significant horizontal movement
            if abs(start_x - end_x) > frame_width * 0.02:
                plt.arrow(start_x, start_y,
                         end_x - start_x, end_y - start_y,
                         head_width=frame_height * 0.01, head_length=frame_width * 0.01,
                         fc=color, ec=color, alpha=0.7)

        plt.xlim(0, frame_width)
        plt.ylim(frame_height, 0)  # Invert Y-axis to match video coordinates
        plt.xlabel("X-coordinate")
        plt.ylabel("Y-coordinate")
        plt.grid(False)
        plt.savefig(os.path.join(self.output_dir, f"{vehicle_type}_lane_changes.png"), dpi=300)
        plt.show()  # Show in notebook

    def _generate_time_based_plot(self, lane_changes, frame_width):
        """
        Generate a time-based plot showing lane changes over time

        Args:
            lane_changes: Dictionary of lane changes by vehicle type
            frame_width: Width of the video frame
        """
        # Determine the total video duration
        max_time = 0
        for vehicle_type, segments in lane_changes.items():
            for _, end_point in segments:
                if end_point[4] > max_time:
                    max_time = end_point[4]

        # Create time groups (e.g., 5-10 second intervals)
        time_interval = self.time_group_seconds
        num_time_groups = int(np.ceil(max_time / time_interval))

        # Create separate time-based plots for each vehicle type
        for vehicle_type, segments in lane_changes.items():
            if not segments:
                continue

            plt.figure(figsize=(15, 10))
            plt.title(f"{vehicle_type.title()} Lane Changes Over Time")

            # Draw lanes
            for i in range(self.lane_count + 1):
                lane_x = int(frame_width * i / self.lane_count)
                plt.axvline(x=lane_x, color='black', linestyle='--', alpha=0.3)

                # Label lanes
                if i < self.lane_count:
                    lane_center = int(frame_width * (i + 0.5) / self.lane_count)
                    plt.text(lane_center, 0, f"Lane {i+1}", ha='center')

            # Define color map for time groups
            cmap = plt.cm.viridis
            norm = plt.Normalize(0, num_time_groups-1)

            # Create legend entries for time groups
            legend_entries = []

            # Process lane changes
            for start_point, end_point in segments:
                start_lane = start_point[3]
                end_lane = end_point[3]
                time_group = int(end_point[4] / time_interval)

                # Skip if no lane change
                if start_lane == end_lane:
                    continue

                # Calculate positions for the arrow
                start_x = frame_width * (start_lane + 0.5) / self.lane_count
                end_x = frame_width * (end_lane + 0.5) / self.lane_count

                # Use time group for y-axis position
                y_pos = time_group * 10 + 20  # Offset to make room for lane labels

                # Get color based on time group
                color = cmap(norm(time_group))

                # Draw the arrow
                plt.arrow(start_x, y_pos,
                         end_x - start_x, 0,  # Horizontal arrow (no y change)
                         head_width=5, head_length=frame_width * 0.02,
                         fc=color, ec=color, alpha=0.7)

                # Add time group to legend if not already present
                time_start = time_group * time_interval
                time_end = min((time_group + 1) * time_interval, max_time)
                time_label = f"{time_start:.1f}-{time_end:.1f}s"

                if time_label not in [l.get_label() for l in legend_entries]:
                    legend_entries.append(plt.Line2D([0], [0], color=color, lw=4, label=time_label))

            plt.xlim(0, frame_width)
            plt.ylim(0, num_time_groups * 10 + 40)  # Adjust y-axis to fit all time groups
            plt.xlabel("Lane Position")
            plt.ylabel("Time Group")
            plt.yticks([])  # Hide y-axis ticks

            # Add legend for time groups
            if legend_entries:
                plt.legend(handles=legend_entries, title="Time Intervals (seconds)",
                          loc='upper right', bbox_to_anchor=(1.15, 1))

            plt.grid(False)
            plt.tight_layout()
            plt.savefig(os.path.join(self.output_dir, f"{vehicle_type}_time_based_lane_changes.png"), dpi=300)
            plt.show()

        # Create a combined time-based plot for all vehicle types
        plt.figure(figsize=(15, 10))
        plt.title("All Vehicles Lane Changes Over Time")

        # Draw lanes
        for i in range(self.lane_count + 1):
            lane_x = int(frame_width * i / self.lane_count)
            plt.axvline(x=lane_x, color='black', linestyle='--', alpha=0.3)

            # Label lanes
            if i < self.lane_count:
                lane_center = int(frame_width * (i + 0.5) / self.lane_count)
                plt.text(lane_center, 0, f"Lane {i+1}", ha='center')

        # Define color map for vehicle types
        vehicle_colors = {'car': 'blue', 'bike': 'green', 'rickshaw': 'orange', 'htv': 'red'}

        # Process lane changes for all vehicle types
        y_offset = 20  # Initial offset for arrows

        # Add legend entries for vehicle types
        vehicle_legend = []
        for vehicle_type, color in vehicle_colors.items():
            vehicle_legend.append(plt.Line2D([0], [0], color=color, lw=4, label=vehicle_type.title()))

        # Time group legends
        time_legend = []

        # Group time segments
        time_grouped_segments = defaultdict(list)

        # Organize all segments by time group
        for vehicle_type, segments in lane_changes.items():
            for start_point, end_point in segments:
                time_group = int(end_point[4] / time_interval)
                if start_point[3] != end_point[3]:  # Only if lane changed
                    time_grouped_segments[time_group].append((vehicle_type, start_point, end_point))

        # Plot each time group
        for time_group in sorted(time_grouped_segments.keys()):
            segments = time_grouped_segments[time_group]
            time_start = time_group * time_interval
            time_end = min((time_group + 1) * time_interval, max_time)

            # Y position for this time group
            y_pos = y_offset + time_group * 15

            # Add time group label
            time_label = f"{time_start:.1f}-{time_end:.1f}s"
            plt.text(10, y_pos - 5, time_label, fontsize=10)

            # Add time group to legend if not already present
            if time_label not in [l.get_label() for l in time_legend]:
                time_legend.append(plt.Line2D([0], [0], color='black', lw=1, label=time_label))

            # Draw arrows for each vehicle in this time group
            for vehicle_type, start_point, end_point in segments:
                start_lane = start_point[3]
                end_lane = end_point[3]

                # Calculate positions for the arrow
                start_x = frame_width * (start_lane + 0.5) / self.lane_count
                end_x = frame_width * (end_lane + 0.5) / self.lane_count

                # Draw the arrow
                plt.arrow(start_x, y_pos,
                         end_x - start_x, 0,  # Horizontal arrow (no y change)
                         head_width=3, head_length=frame_width * 0.02,
                         fc=vehicle_colors[vehicle_type], ec=vehicle_colors[vehicle_type], alpha=0.7)

        plt.xlim(0, frame_width)
        plt.ylim(0, y_offset + len(time_grouped_segments) * 15 + 20)
        plt.xlabel("Lane Position")
        plt.ylabel("Time (seconds)")
        plt.yticks([])  # Hide y-axis ticks

        # Add vehicle type legend
        plt.legend(handles=vehicle_legend, title="Vehicle Types",
                  loc='upper right', bbox_to_anchor=(1.15, 1))

        plt.grid(False)
        plt.tight_layout()
        plt.savefig(os.path.join(self.output_dir, "all_vehicles_time_based_lane_changes.png"), dpi=300)
        plt.show()

    def _download_results(self):
        """
        Compress and download the results folder in Google Colab
        """
        if not IN_COLAB:
            return

        # Zip the output directory
        zip_filename = f"{self.output_dir}.zip"
        !zip -r {zip_filename} {self.output_dir}

        # Download the zipped file
        files.download(zip_filename)
        print(f"Results downloaded as {zip_filename}")

# Now that the class is defined, we can use it
if IN_COLAB:
    # Upload a video file
    print("Please upload your traffic video file:")
    from google.colab import files
    uploaded = files.upload()
    video_path = list(uploaded.keys())[0]
    print(f"Uploaded video: {video_path}")

    # Create and run the analyzer
    analyzer = LaneChangeAnalyzer(
        model_path="yolov8x.pt",
        confidence=0.3,
        lane_count=5,
        time_group_seconds=10  # Group time in 10-second intervals
    )
    analyzer.process_video(
        video_path=video_path,
        sample_rate=2,   # Process every Nth frame
        max_frames=500,  # Process only this many frames (remove for full video)
        display_output=True
    )
```
